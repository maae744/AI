```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
```

```python
file_path = 'C:/AI/iris.data'

#class = Iris Setosa, Iris Versicolour, Iris Virginica
column_names = ["sepal length ", "sepal width ", "petal length ", "petal width ", "class"]          

df = pd.read_csv(file_path, header=None, names=column_names)

df.head(10)
```

```python
X=df.drop('class',axis=1)
X.head()
```

```python
y=df['class']
y.head(10)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
iris_ML = LogisticRegression(random_state=0)
iris_ML.fit(X_train, y_train)

pred_iris_ML = iris_ML.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_iris_ML))
print (confusion_matrix(y_test, pred_iris_ML))
```

```python
iris_ML = LogisticRegression(random_state=0, max_iter=1000)
iris_ML.fit(X_train, y_train)

pred_iris_ML = iris_ML.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_iris_ML))
print (confusion_matrix(y_test, pred_iris_ML))
```
